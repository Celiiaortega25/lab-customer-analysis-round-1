```python
#Standardize header names and Rearrange the columns in the dataframe as needed
```

```python
df1.head()
df1 = df1.rename(columns={'Customer':'customer', 'ST':'state', 'GENDER':'gender', 'Education':'education', 'Customer Lifetime Value':'customer_lifetime_value', 'Income':'income', 'Monthly Premium Auto':'monthly_premium_auto', 'Number of Open Complaints':'number_of_open_complaints', 'Policy Type':'policy_type', 'Vehicle Class': 'vehicle_class', 'Total Claim Amount':'total_claim_amount'})
df1.head()
```

```python
df2.head()
df2 = df2.rename(columns={'Customer':'customer', 'ST':'state', 'GENDER':'gender', 'Education':'education', 'Customer Lifetime Value':'customer_lifetime_value', 'Income':'income', 'Monthly Premium Auto':'monthly_premium_auto', 'Number of Open Complaints':'number_of_open_complaints', 'Policy Type':'policy_type', 'Vehicle Class': 'vehicle_class', 'Total Claim Amount':'total_claim_amount'})
df2 = df2[['customer','state','gender','education','customer_lifetime_value','income','monthly_premium_auto','number_of_open_complaints','policy_type','vehicle_class','total_claim_amount']]
df2.head()
```

```python
df3.head()
df3 = df3.rename(columns={'Customer':'customer', 'State':'state', 'Gender':'gender', 'Education':'education', 'Customer Lifetime Value':'customer_lifetime_value', 'Income':'income', 'Monthly Premium Auto':'monthly_premium_auto', 'Number of Open Complaints':'number_of_open_complaints', 'Policy Type':'policy_type', 'Vehicle Class': 'vehicle_class', 'Total Claim Amount':'total_claim_amount'})
df3 = df3[['customer','state','gender','education','customer_lifetime_value','income','monthly_premium_auto','number_of_open_complaints','policy_type','vehicle_class','total_claim_amount']]
df3.head()
```

```python
#Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 
#100 and change dtype to numerical type. (only 2 of the files need this step.)
```

```python
df1.info()
df1['customer_lifetime_value'] = pd.to_numeric(df1['customer_lifetime_value'], errors='coerce')
df1['customer_lifetime_value'] = df1['customer_lifetime_value'] / 100.00
```

```python
df2.info()
df2['customer_lifetime_value'] = pd.to_numeric(df2['customer_lifetime_value'], errors='coerce')
df2['customer_lifetime_value'] = df2['customer_lifetime_value'] / 100.00
```

```python
#Concatenate the three dataframes
```

```python
df = pd.concat([df1, df2, df3], axis=0)
```
